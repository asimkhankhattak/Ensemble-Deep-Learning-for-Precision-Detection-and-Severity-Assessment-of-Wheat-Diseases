### `Showing Directories`

```python
import os

def show_directories(path, indent=0):
    # Check if the provided path is valid
    if not os.path.exists(path):
        print(f"The path '{path}' does not exist.")
        return

    # Print the current directory
    print(' ' * indent + f"[ðŸ“] {os.path.basename(path)}")

    # Dictionary to track the first file for each file type
    file_types_seen = {}

    # Iterate through all items in the directory
    for item in os.listdir(path):
        item_path = os.path.join(path, item)

        if os.path.isdir(item_path):
            # Recursively display subdirectories
            show_directories(item_path, indent + 4)
        else:
            # Handle files: Extract file extension
            file_extension = os.path.splitext(item)[1].lower()

            # Print only one file for each file type
            if file_extension not in file_types_seen:
                file_types_seen[file_extension] = item
                print(' ' * (indent + 4) + f"  ðŸ“„ {item} (type: {file_extension})")

# Provide the starting path
starting_path = r"/content/YELLOW-RUST-19/YELLOW-RUST-19"
show_directories(starting_path)
```

`Making DataFrame of Data Directories and Data Labels`

```python
# Define dataset path
data_dir = '/content/YELLOW-RUST-19/YELLOW-RUST-19'
classes = os.listdir(data_dir)

# Get all file paths and labels
data = []
labels = []
for cls in classes:
    class_path = os.path.join(data_dir, cls)
    images = os.listdir(class_path)
    for img in images:
        img_path = os.path.join(class_path, img)
        data.append(img_path)
        labels.append(cls)

# Create DataFrame
df = pd.DataFrame({'image_path': data, 'label': labels})
```

```python
df.head()
```

```python
train_df, temp_df = train_test_split(
    df,
    train_size=0.8,
    shuffle=True,
    random_state=42,
    stratify=df['label']
)

valid_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    shuffle=True,
    random_state=42,
    stratify=temp_df['label']
)
```

```python
batch_size = 32
img_size = (224, 224)
channels = 3
img_shape = (img_size[0], img_size[1], channels)
```

```python
tr_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

ts_gen = ImageDataGenerator(rescale=1./255)
```

<div style="background-color:#E6F9E6; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1); border:2px solid #66C2A5; margin-top: 20px;">

    </h1>

```python
train_gen = tr_gen.flow_from_dataframe(
    train_df,
    x_col='image_path',
    y_col='label',
    target_size=img_size,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=True,
    batch_size=batch_size
)

valid_gen = ts_gen.flow_from_dataframe(
    valid_df,
    x_col='image_path',
    y_col='label',
    target_size=img_size,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=True,
    batch_size=batch_size
)

test_gen = ts_gen.flow_from_dataframe(
    test_df,
    x_col='image_path',
    y_col='label',
    target_size=img_size,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=False,
    batch_size=batch_size
)
```

```python
dataset_path = '/content/YELLOW-RUST-19/YELLOW-RUST-19'

# Number of images per category
num_images_per_category = 5

# Supported image file extensions
image_extensions = ['.png', '.jpg', '.jpeg', '.bmp', '.tiff', '.gif']

# Get the list of categories (excluding files like README at the root level)
categories = [cat for cat in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, cat))]

# Iterate over categories and plot images
for category in categories:
    category_path = os.path.join(dataset_path, category)

    # Get image files in the category folder
    images = [f for f in os.listdir(category_path) if os.path.splitext(f)[1].lower() in image_extensions]

    # Limit to the first `num_images_per_category`
    images = images[:num_images_per_category]

    print(f"Category: {category}")
    plt.figure(figsize=(15, 5))

    for i, image_name in enumerate(images):
        image_path = os.path.join(category_path, image_name)
        try:
            image = Image.open(image_path)
            plt.subplot(1, num_images_per_category, i + 1)
            plt.imshow(image)
            plt.axis('off')
            plt.title(f"{category} {i+1}")
        except Exception as e:
            print(f"Error loading image {image_name}: {e}")

    plt.tight_layout()
    plt.show()
```

```python
def build_vgg16_custom_model(input_shape):
    # Input layer
    input_layer = Input(shape=input_shape)

    # Load pre-trained VGG16 model without top layers
    vgg_base = VGG16(weights='imagenet', include_top=False, input_tensor=input_layer)

    # Unfreeze layers after block3_pool
    freeze = True
    for layer in vgg_base.layers:
        if layer.name == 'block3_pool':
            freeze = False
        layer.trainable = not freeze

    # Extract features from the base model
    feature_output = vgg_base.output

    # Reshape the feature maps
    reshaped_output = Reshape((7 * 7, 512))(feature_output)

    # Apply Multi-Head Attention
    attention_layer = MultiHeadAttention(num_heads=8, key_dim=512)
    attention_output = attention_layer(reshaped_output, reshaped_output)
    attention_output = Reshape((7, 7, 512))(attention_output)

    # Global pooling and fully connected layers
    pooled_output = GlobalAveragePooling2D()(attention_output)
    dense_1 = Dense(512, activation='relu')(pooled_output)
    norm_layer = BatchNormalization()(dense_1)
    dropout_layer = Dropout(0.4)(norm_layer)

    # Final classification layer
    final_output = Dense(6, activation='softmax')(dropout_layer)

    # Build the model
    custom_model = Model(inputs=input_layer, outputs=final_output)
    return custom_model

# Model input shape
input_dimensions = (224, 224, 3)

# Initialize and compile the model
final_model = build_vgg16_custom_model(input_dimensions)
```

```python
def create_vgg19_model(input_shape):

    inputs = Input(shape=input_shape)

    base_model = VGG19(weights='imagenet', input_tensor=inputs, include_top=False)

    trainable = False

    for layer in base_model.layers:
        if layer.name == 'block5_conv1':
            trainable = True
        layer.trainable = trainable

    x = base_model.output

    height, width, channels = 7, 7, 512
    x = Reshape((height * width, channels))(x)

    attention_output = MultiHeadAttention(num_heads=6, key_dim=channels)(x, x)

    attention_output = Reshape((height, width, channels))(attention_output)

    x = GaussianNoise(0.45)(attention_output)
    x = GlobalAveragePooling2D()(x)

    x = Dense(512, activation='relu')(x)
    x = BatchNormalization()(x)
    x = GaussianNoise(0.45)(x)
    x = Dropout(0.35)(x)

    outputs = Dense(6, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)

    return model

input_shape = (224, 224, 3)
vgg19_attention_model = create_vgg19_model(input_shape)

vgg19_attention_model.compile(optimizer=Adam(learning_rate=0.0001),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
```

```python
def create_resnet50_model(input_shape):

    inputs = Input(shape=input_shape)

    base_model = ResNet50(weights='imagenet', input_tensor=inputs, include_top=False)


    trainable = False

    for layer in base_model.layers:
        if layer.name == 'conv3_block1_1_conv':
          trainable = True
        layer.trainable = trainable

    x = base_model.output

    height, width, channels = 7, 7, 2048
    x = Reshape((height * width, channels))(x)

    attention_output = MultiHeadAttention(num_heads=8, key_dim=channels)(x, x)

    attention_output = Reshape((height, width, channels))(attention_output)

    x = GaussianNoise(0.25)(attention_output)
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    x = BatchNormalization()(x)
    x = GaussianNoise(0.25)(x)
    x = Dropout(0.25)(x)
    outputs = Dense(6, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)

    return model

input_shape = (224, 224, 3)
ResNet50_attention_model = create_resnet50_model(input_shape)
```
